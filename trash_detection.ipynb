```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
```

Загрузка модели

```python
model = YOLO("model/best.pt")
```

Детекция одного изображения

```python
# путь к изображению
image_path = ("data/Pictures/10.png")

# поддерживаемые форматы
extensions = (".jpg", ".jpeg", ".png")

# загрузка картинки
img = cv2.imread(image_path)

# предсказание
res = model.predict(img, conf=0.05, verbose=False)[0]

print("Найденные объекты:")
for box in res.boxes:
    cls = int(box.cls)
    conf = float(box.conf)
    name = model.names[cls]
    print(f"- {name} ({conf:.2f})")

# визуализация
annotated = res.plot()

# показать через matplotlib
plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Детекция нескольких изображений

```python
# папка с изображениями
folder_path = "data/Pictures"

# поддерживаемые форматы
extensions = (".jpg", ".jpeg", ".png")

for filename in os.listdir(folder_path):
    if filename.lower().endswith(extensions):
        image_path = os.path.join(folder_path, filename)

        # читаем изображение
        img = cv2.imread(image_path)
        res = model.predict(img, conf=0.1, verbose=False)[0]

        print(f"\n==== {filename} ====")
        print("Найденные объекты:")

        for box in res.boxes:
            cls = int(box.cls)
            conf = float(box.conf)
            name = model.names[cls]
            print(f"- {name} ({conf:.2f})")

        # визуализация
        annotated = res.plot()

        # показать результат
        plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```
